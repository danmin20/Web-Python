### "나는 현재 안전한가?"

## (1) 주제 선정 이유

크고 작은 여러 범죄들이 판을 치는 현대 사회에서, 우리는 항상 범죄에 대한 두려움을 안고 살아간다.<br>
범죄는 다양한 장소에서 발생한다. 성별과 나이에 따라서도 범죄율이 다르다.<br>
나에게 어떠한 범죄가 일어날 수 있는지 인지하며 살아간다면,<br>
범죄 상황을 상상해보는 등의 시도로 그에 대한 대처능력이 상승할 수 있지 않을까하는 생각이 들었다.<br>
범죄에 대한 경각심을 더욱 키우고, 그에 대한 대처능력 또한 키우기 위하여,<br>
'현재 나에게 범죄가 일어날 확률'을 알려주는 어플리케이션을 만드는 프로젝트를 구상하게 되었다.

## (2) 가설 정의

경희대학교 국제캠퍼스에 재학 중인 21살 여자 대학생 이정민은 안전하다.

생각보다 범죄에 관련된 자료가 많지 않아 더욱 현실적으로 정확한 결과를 도출해내지 못해 아쉬움이 크다.<br>
실시간으로 접수된 범죄를 집계해주는 사이트가 있다면 거기서 데이터를 가져와 더 나은 프로그램을 만들 수 있지 않을까 한다.

```python
'''
tkinter을 이용하여 GUI 구성
함수의 파라미터 및 기능 수정
'''


# imports
import tkinter as tk
import tkinter.scrolledtext as tkst
from tkinter import Menu
from tkinter import ttk


#최종 함수들

def CrimeArea():
    a=aarea.get()
    
    area_lst=[]
    val=area.values
    
    a=' '+a+' '
    lst=[i for i in area[a]]
    
    summ = lst[0]
    final = lst[:]
    final.remove(final[0])

    for j in final:
        area_lst.append([val[lst.index(j)][0], j])

    result = []
    for i in area_lst:
        result.append([i[0], round(i[1] / summ * 100, 2)])
    return (result)

def CrimePlace():
    p=pplace.get()
    
    val=place.values
    place_lst=[]
    
    if p == '집':
        L1 = [i for i in place['아파트']]
        L2 = [i for i in place['단독주택']]
        lst = [L1[i] + L2[i] for i in range(len(L1))]
    elif p == '길거리':
        lst = [i for i in place['노상']]
    elif p == '상점':
        L1 = [i for i in place['슈퍼마켓']]
        L2 = [i for i in place['편의점']]
        L3 = [i for i in place['상점']]
        lst = [L1[i] + L2[i] + L3[i] for i in range(len(L1))]
    elif p == '숙박업소 또는 목욕탕':
        lst = [i for i in place['숙박업소,목욕탕']]
    elif p == '역 또는 대합실':
        lst = [i for i in place['역,대합실']]
    elif p == '교통수단':
        L1 = [i for i in place['지하철']]
        L2 = [i for i in place['기타교통']]
        lst = [L1 + L2 for i in range(len(L1))]
    elif p == '학교':
        lst = [i for i in place['학교']]
    elif p == '금융기관':
        lst = [i for i in place['금융기관']]
    elif p == '의료기관':
        lst = [i for i in place['의료기관']]
    elif p == '종교기관':
        lst = [i for i in place['종교기관']]
    elif p == '주차장':
        lst = [i for i in place['주차장']]
    elif p == '공중화장실':
        lst = [i for i in place['공중화장실']]
    elif p == 'PC방':
        lst = [i for i in place['PC방']]  
        
    summ = lst[0]
    final = lst[:]
    final.remove(final[0])

    for j in final:
        place_lst.append([val[lst.index(j)][0], j])

    result = []
    for i in place_lst:
        result.append([i[0], round(i[1] / summ * 100, 2)])
    return (result)

def CrimeSexAge():
    s=gender.get()
    ag=age.get()
    lst=[]
    
    if s == '남자':
        if ag <=6:
            val = sexage[' 남 6세이하 '].values
        elif ag <=12:
            val = sexage[' 남 12세이하 '].values
        elif ag <=15:
            val = sexage[' 남 15세이하 '].values
        elif ag <=20:
            val = sexage[' 남 20세이하 '].values
        elif ag <=30:
            val = sexage[' 남 30세이하 '].values
        elif ag <=40:
            val = sexage[' 남 40세이하 '].values
        elif ag <=50:
            val = sexage[' 남 50세이하 '].values
        elif ag <=60:
            val = sexage[' 남 60세이하 '].values
        elif ag >60:
            val = sexage[' 남 60세초과 '].values

    if s == '여자':
        if ag <=6:
            val = sexage[' 여 6세이하 '].values
        elif ag <=12:
            val = sexage[' 여 12세이하 '].values
        elif ag <=15:
            val = sexage[' 여 15세이하 '].values
        elif ag <=20:
            val = sexage[' 여 20세이하 '].values
        elif ag <=30:
            val = sexage[' 여 30세이하 '].values
        elif ag <=40:
            val = sexage[' 여 40세이하 '].values
        elif ag <=50:
            val = sexage[' 여 50세이하 '].values
        elif ag <=60:
            val = sexage[' 여 60세이하 '].values
        elif ag >60:
            val = sexage[' 여 60세초과 '].values

    for i in val:
        n = 0
        for j in i:
            if type(j) == int:
                n += j
        lst.append(n)

    kind = [i for i in sexage[' 2008년 ']]
    num = lst.index(max(lst))

    return (kind[num].strip())

def CrimeTime():
    t=ttime.get()
    
    val = time.values
    time_lst = []

    if t == '00~03시':
        lst = [i for i in time[1]]
    elif t == '03~06시':
        lst = [i for i in time[2]]
    elif t == '06~09시':
        lst = [i for i in time[3]]
    elif t == '09~12시':
        lst = [i for i in time[4]]
    elif t == '12~15시':
        lst = [i for i in time[5]]
    elif t == '15~18시':
        lst = [i for i in time[6]]
    elif t == '18~21시':
        lst = [i for i in time[7]]
    elif t == '21~24시':
        lst = [i for i in time[8]]

    summ = lst[0]
    final = lst[:]
    final.remove(final[0])

    for j in final:
        time_lst.append([val[lst.index(j)][0], j])

    result = []
    for i in time_lst:
        result.append([i[0], round(i[1] / summ * 100, 2)])
    return (result)
        
    
# 입출력

def clickOK():
    text= "Your Area is "+aarea.get()
    text= text +"\nYour Place is "+pplace.get()
    text = text + "\nYour gender is " + gender.get()
    text = text + "\nYou are " + str(age.get()) + " years old.\n"
    text = text + "\n\n"
    
    Aresult = CrimeArea()
    Presult = CrimePlace()
    SAresult = CrimeSexAge()
    Tresult = CrimeTime()

    text = str(age.get())+'세 '+gender.get()+"가 가장 많이 겪는 범죄는 [" + SAresult + "]입니다.\n\n"
    text = text + '\n'+aarea.get()+ "에서의 범죄율은 다음과 같습니다.\n\n"
    for i in Aresult:
        text = text + str(i[0]) + ':' + str(i[1]) + '%\n'
    text = text + '\n\n'+pplace.get()+"에서의 범죄율은 다음과 같습니다.\n\n"
    for i in Presult:
        text = text + str(i[0]) + ':' + str(i[1]) + '%\n'
    text=text+"\n\n"+ttime.get()+"의 범죄율은 다음과 같습니다.\n\n"
    for i in Tresult:
        text = text +str(i[0])+':'+str(i[1])+'%\n'
    
    #text를 scrolledtext에 넣기
    scrt.insert(tk.INSERT, text)
    scrt.see(tk.END)
    
#exit메뉴
def clickExit():
    win.quit()
    win.destroy()
    exit()

#GUI
if __name__=='__main__':
    win=tk.Tk()
    win.title("현재 나에게 범죄가 일어날 확률")
    
    labelArea=ttk.Label(win,text="Area")
    labelArea.grid(column=0, row=0)
    
    labelPlace=ttk.Label(win,text="Place")
    labelPlace.grid(column=1,row=0)
    
    labelGender = ttk.Label(win, text="Gender")
    labelGender.grid(column=2, row=0)
    
    labelAge = ttk.Label(win, text="Age")
    labelAge.grid(column=3, row=0)
    
    labelTime=ttk.Label(win,text="Time")
    labelTime.grid(column=4,row=0)
    
    aarea=tk.StringVar()
    areaCombo=ttk.Combobox(win,width=10,textvariable=aarea)
    areaCombo['values']=('서울', '부산', '대구', '인천', '광주', '대전', '울산', '부천', '수원', '성남', '전주', '안양', '청주', '마산', '창원', '광명', '포항',
                 '안산', '진주', '고양', '제주', '목포', '의정부', '익산', '군산', '구미', '천안', '여수', '춘천', '원주', '평택', '경주', '김해', '순천',
                 '군포', '남양주', '강릉', '충주', '안동', '경산', '아산', '거제', '김천', '정읍', '용입', '시흥', '파주', '양산', '이천', '구리', '서산',
                 '제천', '논산', '기타도시', '도시이외')
    areaCombo.grid(column=0,row=1)
    areaCombo.current(0)
    
    pplace=tk.StringVar()
    placeCombo=ttk.Combobox(win, width=17, textvariable=pplace)
    placeCombo['values']=('집','길거리','상점','숙박업소 또는 목욕탕','역 또는 대합실','교통수단','학교','금융기관','의료기관','종교기관','주차장','공중화장실','PC방')
    placeCombo.grid(column=1,row=1)
    placeCombo.current(0)
    
    gender = tk.StringVar()
    genderCombo = ttk.Combobox(win, width=6, textvariable=gender)
    genderCombo['values'] = ("남자", "여자")
    genderCombo.grid(column=2, row=1)
    genderCombo.current(0)

    age = tk.IntVar()
    ageEntered = ttk.Entry(win, width=3, textvariable=age)
    ageEntered.grid(column=3, row=1)
    
    ttime=tk.StringVar()
    timeCombo=ttk.Combobox(win, width=10, textvariable=ttime)
    timeCombo['values']=('00~03시','03~06시','06~09시','09~12시','12~15시','15~18시','18~21시','21~24시')
    timeCombo.grid(column=4, row=1)
    timeCombo.current(0)

    action = ttk.Button(win, text="OK", command=clickOK)
    action.grid(column=5, row=1)

    scrt = tkst.ScrolledText(win, width=70, height=50, wrap=tk.WORD)
    scrt.grid(column=0, row=2, columnspan=5)
    scrt.focus_set()
    
    menuBar = Menu(win)
    win.config(menu=menuBar)

    fileMenu = Menu(menuBar, tearoff=0)
    fileMenu.add_command(label="New")
    fileMenu.add_separator()
    fileMenu.add_command(label="Exit", command=clickExit)
    menuBar.add_cascade(label="File", menu=fileMenu)

    win.resizable(0, 0)
    win.mainloop()
```
